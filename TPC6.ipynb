```python
# TabMeteo = [(Data,TempMin,TempMax,Precipitacao)]
    # Data = (Int,Int,Int)
    # TempMin = Float
    # TempMax = Float
    # Precipitacao = Float
import matplotlib.pyplot as plt

def medias(tabMeteo):
    res = []
    for dia in tabMeteo:
        res.append((dia[0],(dia[1]+dia[2])/2))
    return res

def guardaTabMeteo(t, fnome):
    file=open(fnome,"w", encoding='utf-8')
    res=""
    for dia in t:
        file.write(f"{dia[0][0]};{dia[0][1]};{dia[0][2]};{dia[1]};{dia[2]};{dia[3]}\n")
    file.close()
    return

def carregaTabMeteo(fnome):
    res = []
    file=open(fnome,"r",encoding='utf-8')
    for linha in file:
        campos=linha.split(";")
        res.append(((int(campos[0]),int(campos[1]),int(campos[2])),float(campos[3]),float(campos[4]),float(campos[5])))
    file.close()
    return res

def minMin(tabMeteo):
    menor=tabMeteo[0][1]
    for reg in tabMeteo[1:]:
        if reg[1]<menor:
            menor=reg[1]
    return menor

def amplTerm(tabMeteo):
    res = []
    for reg in tabMeteo:
        amp=reg[2]-reg[1]
        res.append((reg[0],amp))
    return res 

def maxChuva(tabMeteo):
    max_prec=tabMeteo[0][3]
    for reg in tabMeteo[1:]:
        if reg[3]>max_prec:
            max_prec=reg[3]
            max_data=reg[0]
    return (max_data, max_prec)

def diasChuvosos(tabMeteo, p):
    res=[]
    for reg in tabMeteo:
        if reg[3]>p:
            res.append((reg[0],reg[3]))
    return res

def maxPeriodoCalor(tabMeteo, p):
    período_atual=0
    período_maximo=0
    for reg in tabMeteo:
        if reg[3]<p:
            período_atual=período_atual+1
        else:
            if período_atual>período_maximo:
                período_maximo=período_atual
            período_atual=0
    if período_atual>período_maximo:
        período_maximo=período_atual
    return período_maximo


def extraiTMin(t):
    res=[]
    for _,tmin,_,_ in t:
        res.append(tmin)
    return res
def extraiTMax(t):
    res=[]
    for _,_,tmax,_ in t:
        res.append(tmax)
    return res

def extraiPrecip(t):
    res=[]
    for _,_,_,precip in t:
        res.append(precip)
    return res


def grafTabMeteo(t):
    # Tmin
    x1 = list(range(1,len(t)+1))
    y1 = extraiTMin(t)
    plt.plot(x1,y1,label="Temperatura Mínima")
    #Tmax
    x2=list(range(1,len(t)+1))
    y2=extraiTMax(t)
    plt.plot(x2,y2,label="Temperatura Maxima")

    #Precip
    x3=list(range(1,len(t)+1))
    y3=extraiPrecip(t)
    plt.plot(x3,y3,label="Precipitação")


    plt.title('Meterologia')
    plt.legend()
    plt.show()
    return


def menu():
    print(""" 
    -----------------------------------------------------------------------------------
    Menu de opções:
    (1) Consultar temperatura média de cada dia
    (2) Consultar a menor temperatura da tabela de registos
    (3) Consultar a amplitude térmica de cada dia
    (4) Consultar o dia com maior nível de precipitação
    (5) Consultar os dias em que a precipitação foi maior que um certo valor limite
    (6) Consultar o nº de dias consecutivos com precipitação inferior a um valor limite
    (7) Criar gráfico alusivo à tabela de registos
    (8) Guardar a tabela de registos meteorológicos em ficheiro
    (9) Carregar tabela de registos meteorológicos dum ficheiro
    (0) Sair
    -----------------------------------------------------------------------------------
    """)
#Assumindo que os dias estão ordenados por ordem crescente.
tabMeteo = [((2022,1,20), 2, 16, 0),((2022,1,21), 1, 13, 0.2), ((2022,1,22), 7, 17, 0.01), ((2022,1,23),11, 25,0), ((2022,1,24),8,14,3.2), ((2022,1,25), 8, 11, 2.1), ((2022,1,26), 12, 21, 0.3)]
variável=True
print(tabMeteo)
while variável==True:
    menu()
    resp=int(input("Qual o número da opção que pretende utilizar?"))
    if resp==1:
        print(medias(tabMeteo))
    elif resp==2:
        print(minMin(tabMeteo))
    elif resp==3:
        print(amplTerm(tabMeteo))
    elif resp==4:
        print(maxChuva(tabMeteo))
    elif resp==5:
        valor=float(input("Qual o valor de precipitação que pretende usar como valor limite?"))
        if valor<0:
            print("Erro! Não podes inserir valores negativos para a precipitação!")
        else:
            print(diasChuvosos(tabMeteo,valor))
    elif resp==6:
        valor=float(input("Qual o valor de precipitação que pretende usar como valor limite?"))
        if valor<0:
            print("Erro! Não podes inserir valores negativos para a precipitação!")
        else:
            print(maxPeriodoCalor(tabMeteo,valor))
    elif resp==7:
        grafTabMeteo(tabMeteo)
    elif resp==8:
        nome=input("Qual o nome do registo?(A,B,C,D...ou 1,2,3...)").strip().upper()
        if len(nome)>1:
            print("Nome inválido para o registo! Siga as instruções.")
        else:
            guardaTabMeteo(tabMeteo, f"registo{nome}.txt")
            print("Registo guardado com sucesso")
    elif resp==9:
        nomef=input("Digite o nome do ficheiro de onde quer carregar o registo (A,B,C,D...ou 1,2,3...):")
        if len(nomef)>1:
            print("Nome inválido para o registo ou não encontrado! Certifique-se que coloca um número ou uma letra.")
        else:
            reg=(carregaTabMeteo(f"registo{nomef}.txt"))
            print(reg)
    elif resp==0:
        print(f"Saiu do menu de opções!")
        variável=False
    else:
        print("Número inválido! Escolha outra opção.")


    
```
