```python
def mmc(a, b):
    if a < b:
        a, b = b, a
    mult = a
    while mult % b != 0:
        mult += a
    return mult

print(mmc(12, 16))
```

```python
def f2c(fahrenheit):
    celsius_dict = {}
    for key, f in fahrenheit.items():
        c = (f - 32) / 1.8
        celsius_dict[key] = (f, c)
    return celsius_dict

fahrenheit = {'t1': -30, 't2': -20, 't3': -10, 't4': 0}
print(f2c(fahrenheit))
```

```python
def reduxInt(n):
    while n >= 10:
        total = 0
        while n > 0:
            total += n % 10
            n //= 10
        n = total
    return n

print(reduxInt(9875))
```

```python
def existe_medico(hosp, unihosp,nmedico):
    res=False
    for unidade in hosp:
        if unidade[0]==unihosp:
            for med in unidade[1]:
                if med[1]==nmedico:
                    res=True
    return res

print(existe_medico(Hospital, "Unidade de Cardiologia", "Dra. Maria"))
print(existe_medico(Hospital, "Unidade de Cardiologia", "Dr. Pedro"))
```

```python
def adicionar_medico(hosp, unihosp, nmedico):
    novo_hosp = []
    for unidade in hosp:
        if unidade[0] == unihosp:
            medicos = unidade[1]
            existe = False
            for med in medicos:
                if med[1]==nmedico:
                    existe=True

            if not existe:
                novo_id =len(medicos)+1
                medicos =medicos+[(novo_id, nmedico)]
            novo_hosp.append((unidade[0], medicos, unidade[2]))
        else:
            novo_hosp.append(unidade)

    return novo_hosp

print(adicionar_medico(Hospital, "Unidade de Cardiologia", "Dra.Paula"))
```

```python
def ocupar_cama(hosp, unihosp, idcama):
    for unidade in hosp:
        if unidade[0] == unihosp:
            novas_camas=[]
            for cama in unidade[2]:
                if cama[0] == idcama and cama[1]==False:
                    novas_camas.append((cama[0], True))
                else:
                    novas_camas.append(cama)
            unidade[2][:]=novas_camas
    return hosp
    

print(ocupar_cama(Hospital,"Unidade de Cardiologia", 101))
```

```python
def guardar_hospital(hosp, nomef):
    file = open(nomef, "w", encoding="utf-8")
    res = ""
    for unidade in hosp:
        res += f"{unidade[0]}\n"
        for med in unidade[1]:
            res += f"Medico:{med[0]};{med[1]}\n"
        for cama in unidade[2]:
            estado = "Ocupada" if cama[1] else "Livre"
            res += f"Cama:{cama[0]};{estado}\n"
        res += "\n"
    file.write(res)
    file.close()

guardar_hospital(Hospital, "hospital.txt")
```

```python
def camas_ocupadas(hosp):
    total = 0
    ocupadas = 0
    for unidade in hosp:
        for cama in unidade[2]:
            total += 1
            if cama[1]:
                ocupadas += 1
    if total == 0:
        return 0
    return (ocupadas / total) * 100

print(camas_ocupadas(Hospital))
```

```python
def unidades_ocupacao_critica(hosp):
    res = []
    for unidade in hosp:
        camas = unidade[2]
        total = len(camas)
        if total == 0:
            percent = 0
        else:
            ocupadas = sum(1 for cama in camas if cama[1])
            percent = (ocupadas / total) * 100
        if percent >= 80:
            res.append((unidade[0], percent))
    return res

print(unidades_ocupacao_critica(Hospital))
```
