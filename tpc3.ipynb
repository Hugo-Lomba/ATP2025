```python
import random

def menu():
    print ("""
    ----------------------------------------------
    Lista de opções:
    -(1)Criar Lista
    -(2)Ler Lista
    -(3)Soma
    -(4)Média
    -(5)Maior
    -(6)Menor
    -(7)estaOrdenada por ordem crescente
    -(8)estaOredenada por ordem decrescente
    -(9)Procura um elemento
    -(0)Sair
    --------------------------------------------""")
    return

def criaLista(N):
    res=[]
    i=1
    while i<=N:
        res.append(random.randint(1,100))
        i=i+1
    return res

def lêLista(N):
    res=[]
    i=1
    while i<=N:
        x=int(input(f"Introduza o {i}º número:"))
        res.append(x)
        i=i+1
    return res

def somaLista(lista):
    res=0
    for n in lista:
        res=n+res 
    return res

def mediaLista(lista):
    res=0
    for n in lista:
        res=n+res
    return res/len(lista)

def maiorLista(lista):
    res=lista[0]
    for elem in lista[1:]:
        if elem>res:
            res=elem
    return res

def menorLista(lista):
    res=lista[0]
    for elem in lista[1:]:
        if elem<res:
            res=elem
    return res

def estaOrdenadaC(lista):
    i = 0
    ordenada=True
    while i < len(lista) - 1 and ordenada==True:
        if lista[i] > lista[i+1]:
            ordenada=False
        i = i + 1
    if ordenada==True:
        return "Sim"
    else:
        return "Não"
    

def estaOrdenadaD(lista):
    i=0
    ordenadad=True
    while i < len(lista) - 1 and ordenadad==True:
        if lista[i] < lista[i+1]:
            ordenadad=False
        i = i + 1
    if ordenadad==True:
        return "Sim"
    else:
        return "Não"

def indicede(lista, elem):
    i=0
    encontrado=False
    while i<len(lista) and encontrado==False:
        if lista[i]==elem:
            encontrado=True
        i=i+1
    if encontrado==False:
        return "-1"
    else:
        return i-1
lista_momento=[]

variavel=True
while variavel==True:
    menu()
    r=int(input("Digite um número de acordo com a opção que quer executar."))
    if r==1:
        lista_momento=criaLista(5)
        print(lista_momento)
    elif r==2:
        lista_momento=lêLista(5)
        print(lista_momento)
    elif r>=3 and r<=9:
        if lista_momento==[]:
            print("A lista está vazia! Deve selecionar inicialmente a opção 1 ou 2 para obter uma lista!")
        else:
            if r==3:
                print(somaLista(lista_momento))
            elif r==4:
                print(mediaLista(lista_momento))
            elif r==5:
                print(maiorLista(lista_momento))
            elif r==6:
                print(menorLista(lista_momento))
            elif r==7:
                print(estaOrdenadaC(lista_momento))
            elif r==8:
                print(estaOrdenadaD(lista_momento))
            else:
                x=int(input("Qual o número inteiro que quer procurar na lista?"))
                print(indicede(lista_momento,x))
    elif r==0:
        print(f"Saiu do programa! A lista final é {lista_momento}")
        variavel=False
    else:
        print("Número inválido! Escolha outra opção")










```
