```python
def menu():
    print(""" 
    -----------------------------------------------------------------------------------
    Menu de opções:
    (1) Criar e inserir sala
    (2) Remover sala
    (3) Listar salas
    (4) Consultar lugares dísponíveis em todas as salas
    (5) Consulta sala em específico 
    (6) Verificar se o lugar está ocupado
    (7) Compra bilhte e assegura um lugar
    (8) Faz devolução do bilhte e liberta lugar
    (9) Reset
    (0) Sair
    -----------------------------------------------------------------------------------
    """)

def criaSala(listas, nomes, lugares):
    existe = False
    for s in listas:
        if s[2] == nomes:
            existe = True
    if existe:
        print(f"A sala com o filme '{nomes}' já existe!")
    else:
        nova_sala = [lugares, [], nomes]
        listas.append(nova_sala)
        print(f"Sala criada para o filme '{nomes}' com {lugares} lugares.")
    return listas
    


def listar( listas ):
    for s in listas:
        print(s)
    return

def disponivel( listas, nomes, lugar ):
    i=0
    res=-1
    while i<len(listas) and res==-1:#Garanto que sai do ciclo assim que encontra o nome do filme
        if listas[i][2]==nomes:
            res=i
        i=i+1
    return lugar not in listas[res][1] and 0<lugar<=listas[res][0] #disponivel se o lugar nao estiver na lista dos ocupados


def vendebilhete( listas, nomes, lugar):
    res=-1
    i=0
    while i<len(listas) and res==-1:
        if listas[i][2]==nomes:    
            res=i
        i=i+1  
    if lugar not in listas[res][1] and 0<lugar<=listas[res][0] :
        listas[res][1].append(lugar)
        print(f"Bilhete vendido para o lugar {lugar} no filme {nomes}.")
    else:
        print(f"O lugar {lugar} já está ocupado ou não pode ser escolhido.")
    return listas



def listardisponibilidades( listas ):
    i=0
    while i<len(listas):
        lugares_disponíveis=listas[i][0]-len(listas[i][1])
        print(f"Na sala com o filme {listas[i][2]} existem {lugares_disponíveis} lugares disponíveis!")
        i=i+1
    return


def listarS(listas, nomes):
    i=0
    res=-1
    while i<len(listas) and res==-1:
        if listas[i][2]==nomes:
            res=i
        i=i+1
    if res == -1:
        print(f"A sala com o filme {nomes} não existe!")
    else:
        print(listas[res])

    return


def libertaLugar(listas, nomes, lugar):#Devolução de bilhete
    i=0
    res=True
    while i<len(listas) and res==True:
        if listas[i][2]==nomes:
            if lugar in listas[i][1]:
                nova_lista_vendidos=[]
                for l in listas[i][1]:
                    if l!=lugar:
                        nova_lista_vendidos.append(l)
                listas[i][1] = nova_lista_vendidos
                res=False
                print(f"Lugar {lugar} libertado na sala do filme {nomes}.")
            else:
                print(f"O lugar {lugar} não está ocupado na sala do filme {nomes}.")
        i=i+1
    return listas

def removeSala(listas, nomes):
    i = 0
    encontrada = False
    while i < len(listas) and encontrada == False:
        if listas[i][2] == nomes:
            encontrada = True
            if listas[i][1] == []:
                nova_lista_salas = []
                for j in range(len(listas)):
                    if j != i:
                        nova_lista_salas.append(listas[j])
                listas = nova_lista_salas
                print(f"A sala do filme '{nomes}' foi removida!")
            else:
                print(f"Não é possível remover a sala do filme '{nomes}': ainda tem lugares ocupados.")
        i = i + 1

    if encontrada == False:
        print(f"A sala do filme '{nomes}' não existe.")

    return listas


cinema=[]
variavel=True
while variavel==True:
    menu()
    r=int(input("Digite um número de acordo com a opção que quer executar."))
    if r==1:
        n=int(input("Quantos lugares terá a sala no total?"))
        nome1=input("Qual o filme que será exibido na sala?")
        cinema=criaSala(cinema, nome1,n)
    elif r>=2 and r<=9:
        if cinema==[]:
            print("Neste momento não existe nenhuma sala de cinema criada...")
        else:
            if r==9:
                cinema=[]
                print("Fizeste um reset...As salas de cinema criadas até ao momento foram removidas. Não existem salas no cinema!")
            elif r==2:
                nome1=input("Qual o nome do filme que corresponde à sala que quer remover?")
                cinema=removeSala(cinema, nome1)
            elif r==3:
                listar(cinema)
            elif r==4:
                listardisponibilidades(cinema)
            elif r==5:
                nome1=input("Digite o nome do filme que está a ser exibido na sala que quer consultar.")
                listarS(cinema, nome1)
            elif 6<=r<=8:
                nome1=input("Digite o nome do filme:")
                i = 0
                encontrado = False
                while i < len(cinema) and encontrado == False:
                    if cinema[i][2] == nome1:
                        encontrado = True
                        if r == 6:
                            n = int(input("Qual o lugar que pretende ver se está disponível? "))
                            print(disponivel(cinema, nome1, n))
                        elif r == 7:
                            n = int(input("Quer ficar em que lugar? "))
                            cinema = vendebilhete(cinema, nome1, n)
                        else:
                            print("Devolução do bilhete")
                            n = int(input("Qual é o lugar marcado no bilhete que quer devolver? "))
                            cinema = libertaLugar(cinema, nome1, n)
                    i = i + 1
                if encontrado == False:
                    print("Esse filme não foi encontrado!")
    elif r==0:
        print(f"Saiu do menu de opções do cinema.")
        variavel=False
    else:
        print("Número inválido! Escolha outra opção.")
```
