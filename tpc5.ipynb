```python
def menu():
    print(""" 
    -----------------------------------------------------------------------------------
    Menu de opções:
    (1) Criar turma
    (2) Inserir aluno na turma
    (3) Listar a turma
    (4) Consultar aluno por id
    (5) Remover aluno da turma
    (6) Ver qual a melhor nota na turma (num dado parâmetro)
    (7) Alunos reprovados no teste
    (8) Guardar a turma em ficheiro
    (9) Carregar turma dum ficheiro
    (0) Sair
    -----------------------------------------------------------------------------------
    """)

def criarTurma():
    turma=[]
    print("Acabaste de criar uma turma! Agora já podes inserir alunos na mesma!")
    return turma

def verificaçãoaluno(listat,id):
    i=0
    existe=False
    while i<len(listat) and existe==False:
        if listat[i][1]==id:
            existe=True
        i=i+1
    return existe
            
def inseriraluno(listat,novo_aluno):
    listat.append(novo_aluno)
    return listat
    
def listarturma(listat):
    for elem in listat:
        print(elem)
    return

def listarA(listat, id):
    i=0
    res=-1
    while i<len(listat) and res==-1:
        if listat[i][1]==id:
            res=i
        i=i+1
    if res == -1:
        print(f"O aluno com id {id} não existe!")
    else:
        print(listat[res])
    return

def removeA(listat,id):
    i = 0
    encontrado=False
    while i < len(listat) and encontrado==False:
        if listat[i][1] == id:
            encontrado=True
            nova_lista_alunos = []
            for j in range(len(listat)):
                if j != i:
                    nova_lista_alunos.append(listat[j])
            listat = nova_lista_alunos
            print(f"O aluno com id {id} foi removido!")           
        i=i+ 1
    if encontrado==False:
        print(f"O aluno com id {id} não existe.")
    return listat

def melhornota(listat,indice):
    melhor = turma[0]
    for aluno in listat:
        if aluno[2][indice] > melhor[2][indice]:
            melhor = aluno
    print(f"O melhor aluno é {melhor[0]} (ID: {melhor[1]}) com a nota {melhor[2][indice]}.")
    return

def reprovados(listat):
    reprov=[]
    for aluno in listat:
        if aluno[2][2]<9.5:
            reprov.append(aluno)
    if reprov==[]:
        print("Nenhum aluno reprovou no teste.")
    else:
        print(reprov)
    return

def guardarTurma(listat,nomef):
    file=open(nomef,"w")
    res=""
    for aluno in listat:
        nota1,nota2,nota3=aluno[2]
        res=res+str(aluno[0])+";"+str(aluno[1])+";"+str(nota1)+";"+str(nota2)+";"+str(nota3)
        res=res+"\n"
    file.write(res)
    file.close()
    return

def carregarf(nomef):
    listat=[]
    file=open(nomef, "r")
    text=file.read()
    aluno_text=text.split("\n")
    for a_text in aluno_text[:-1]:#no fim há uma sring vazia ""
        parâmetros_text=a_text.split(";")
        nome=parâmetros_text[0]
        id=parâmetros_text[1]
        lista_notas=[float(parâmetros_text[2]),float(parâmetros_text[3]),float(parâmetros_text[4])]
        aluno=(nome,id,lista_notas)
        listat.append(aluno)
    file.close()
    return listat






aluno1=("Manuel", "A110406", [12.0, 15.0, 18.0])
aluno2=("Simao", "A110321", [11.0,8.0,20.0])
turma=[aluno1,aluno2]
variável=True
while variável==True:
    menu()
    resp=int(input("Qual o número da opção que pretende utilizar?"))
    if resp==1:
        if len(turma)>0:
            r=(input("A turma já tem alunos! Ao criar turma os dados desses alunos serão removidos! Se quiser avançar digite 1, senão digite 2."))
            if r=="1":
                turma=criarTurma()
            elif r=="2":
                print("Não criou uma nova turma!")
            else:
                print("Erro!")
        else:
            turma=criarTurma()
    elif resp==2:
        nome=input("Qual o nome do aluno?")
        id=input("Qual o id do aluno?").strip().upper()
        if verificaçãoaluno(turma, id)==False:
            nota1=float(input("Qual a nota que o aluno obteve no TPC?"))
            nota2=float(input("Qual a nota que o aluno obteve no Projeto?"))
            nota3=float(input("Qual a nota que o aluno obteve no Teste?"))
            if 0<=nota1<=20 and 0<=nota2<=20 and 0<=nota3<=20:
                novo_aluno=(nome,id,[nota1,nota2,nota3])
                turma=inseriraluno(turma, novo_aluno)
                print(f"O aluno {nome} com o id {id} foi inserido na turma!")
            else:
                print("As notas só podem ser valores inteiros entre 0 e 20.")
        else:
            print("Já existe um aluno com esse id!")
    elif 3<=resp<=8:
        if turma==[]:
            print("Ainda não existem alunos na turma para poderes executar estas opções.")
        else:
            if resp==3:
                listarturma(turma)
            elif resp==4:
                id=input("Qual o id do aluno que quer procurar").strip().upper()
                listarA(turma,id)
            elif resp==5:
                id=input("Qual o id do aluno que quer remover").strip().upper()
                turma=removeA(turma,id)
            elif resp==6:
                    print("Escolha o tipo de nota:")
                    print("(1) TPC")
                    print("(2) Projeto")
                    print("(3) Teste")
                    opção=input("Opção:")
                    if opção not in ["1","2","3"]:
                        print("Opção inválida!")
                    else:
                        indice=int(opção)-1
                        (melhornota(turma,indice))
            elif resp==7:
                reprovados(turma)
            else:
                nome=input("Qual o nome da turma?(A,B,C,D...ou 1,2,3...)").strip().upper()
                if len(nome)>1:
                    print("Nome inválido para a turma! Siga as instruções.")
                else:
                    guardarTurma(turma, f"turma{nome}.txt")
                    print("Turma guardada com sucesso")
    elif resp==9:
        nomef=input("Digite o nome do ficheiro de onde quer carregar a turma:")
        turma=(carregarf(nomef))
        print(turma)

        
    elif resp==0:
        print(f"Saiu do menu de opções!")
        variável=False
    else:
        print("Número inválido! Escolha outra opção.")




    
```
